```python
nfl_data.columns
```

```python
missing_values = nfl_data.isnull().sum()
total_missing = missing_values.sum()
total_missing
```

```python
total = nfl_data.shape
total
```

```python
total_cells = np.product(nfl_data.shape)
total_cells
```

```python
missing_percent = (total_missing/total_cells) * 100
missing_percent
```
