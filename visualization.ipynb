# Line Chart

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
spotify = pd.read_csv("spotify.csv", index_col="Date", parse_dates=True)
spotify
```

```python
plt.figure(figsize=(14, 6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify)
```

```python
plt.figure(figsize=(14, 6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

sns.lineplot(data=spotify["Shape of You"], label="Shape of You")
sns.lineplot(data=spotify["Despacito"], label="Despacito")
plt.xlabel("date")
```

# Bar Charts and Heatmaps

```python
flight_data = pd.read_csv("flight_delays.csv", index_col="Month")
flight_data
```

```python
plt.figure(figsize=(10, 6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel("Arrival delay in minutes")
```

```python
plt.figure(figsize=(14,10))
plt.title("Average Arrival Delay for Each Airline, by Month")

sns.heatmap(data=flight_data, annot=True)
plt.xlabel("Airline")
```

# Scatter Plots

```python
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```
